```python
# ------------------------------
# üì¶ ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò (ÏóÜÏúºÎ©¥ ÏÑ§Ïπò)
# ------------------------------
try:
    import pandas as pd
    import seaborn as sns
    import matplotlib.pyplot as plt
except ModuleNotFoundError:
    import os
    os.system('pip install pandas matplotlib seaborn')
    import pandas as pd
    import seaborn as sns
    import matplotlib.pyplot as plt

# ------------------------------
# ‚úÖ Windows ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï
# ------------------------------
plt.rcParams['font.family'] = 'Malgun Gothic'  # ÏúàÎèÑÏö∞ ÌïúÍ∏Ä Ìè∞Ìä∏
plt.rcParams['axes.unicode_minus'] = False     # ÎßàÏù¥ÎÑàÏä§ Í∏∞Ìò∏ Íπ®Ïßê Î∞©ÏßÄ

# ------------------------------
# üìÅ Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
# ------------------------------
file_path = r"C:\Users\owner\Desktop\LG Aimers\open\train\processed_with_weekday_holiday.csv"
df = pd.read_csv(file_path, parse_dates=["ÏòÅÏóÖÏùºÏûê"])

# ------------------------------
# üìä ÏöîÏùº / Í≥µÌú¥ÏùºÎ≥Ñ ÌèâÍ∑† Îß§Ï∂ú Í≥ÑÏÇ∞
# ------------------------------
weekday_order = ['Ïõî', 'Ìôî', 'Ïàò', 'Î™©', 'Í∏à', 'ÌÜ†', 'Ïùº']

# ÏöîÏùºÎ≥Ñ ÌèâÍ∑† Îß§Ï∂ú
weekday_avg = df.groupby('ÏöîÏùº')['Îß§Ï∂úÏàòÎüâ'].mean().reindex(weekday_order)

# Í≥µÌú¥Ïùº Ïó¨Î∂ÄÎ≥Ñ ÌèâÍ∑† Îß§Ï∂ú (0=ÎπÑÍ≥µÌú¥Ïùº, 1=Í≥µÌú¥Ïùº)
holiday_avg = df.groupby('Í≥µÌú¥ÏùºÏó¨Î∂Ä')['Îß§Ï∂úÏàòÎüâ'].mean().rename({0: 'ÎπÑÍ≥µÌú¥Ïùº', 1: 'Í≥µÌú¥Ïùº'})

# ------------------------------
# üìà ÏöîÏùºÎ≥Ñ ÌèâÍ∑† Îß§Ï∂ú ÏãúÍ∞ÅÌôî
# ------------------------------
plt.figure(figsize=(8, 5))
sns.barplot(x=weekday_avg.index, y=weekday_avg.values, palette='Blues_d')
plt.title('ÏöîÏùºÎ≥Ñ ÌèâÍ∑† Îß§Ï∂úÏàòÎüâ', fontsize=14)
plt.xlabel('ÏöîÏùº')
plt.ylabel('ÌèâÍ∑† Îß§Ï∂úÏàòÎüâ')
plt.tight_layout()
plt.show()

# ------------------------------
# üìà Í≥µÌú¥Ïùº Ïó¨Î∂Ä ÏãúÍ∞ÅÌôî
# ------------------------------
plt.figure(figsize=(6, 4))
sns.barplot(x=holiday_avg.index, y=holiday_avg.values, palette='Set2')
plt.title('Í≥µÌú¥Ïùº Ïó¨Î∂ÄÏóê Îî∞Î•∏ ÌèâÍ∑† Îß§Ï∂úÏàòÎüâ', fontsize=14)
plt.xlabel('Í≥µÌú¥Ïùº Ïó¨Î∂Ä')
plt.ylabel('ÌèâÍ∑† Îß§Ï∂úÏàòÎüâ')
plt.tight_layout()
plt.show()
```
