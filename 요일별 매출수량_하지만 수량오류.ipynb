```python
# ------------------------------
# 📦 라이브러리 설치 (없으면 설치)
# ------------------------------
try:
    import pandas as pd
    import seaborn as sns
    import matplotlib.pyplot as plt
except ModuleNotFoundError:
    import os
    os.system('pip install pandas matplotlib seaborn')
    import pandas as pd
    import seaborn as sns
    import matplotlib.pyplot as plt

# ------------------------------
# ✅ Windows 한글 폰트 설정
# ------------------------------
plt.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우 한글 폰트
plt.rcParams['axes.unicode_minus'] = False     # 마이너스 기호 깨짐 방지

# ------------------------------
# 📁 데이터 불러오기
# ------------------------------
file_path = r"C:\Users\owner\Desktop\LG Aimers\open\train\processed_with_weekday_holiday.csv"
df = pd.read_csv(file_path, parse_dates=["영업일자"])

# ------------------------------
# 📊 요일 / 공휴일별 평균 매출 계산
# ------------------------------
weekday_order = ['월', '화', '수', '목', '금', '토', '일']

# 요일별 평균 매출
weekday_avg = df.groupby('요일')['매출수량'].mean().reindex(weekday_order)

# 공휴일 여부별 평균 매출 (0=비공휴일, 1=공휴일)
holiday_avg = df.groupby('공휴일여부')['매출수량'].mean().rename({0: '비공휴일', 1: '공휴일'})

# ------------------------------
# 📈 요일별 평균 매출 시각화
# ------------------------------
plt.figure(figsize=(8, 5))
sns.barplot(x=weekday_avg.index, y=weekday_avg.values, palette='Blues_d')
plt.title('요일별 평균 매출수량', fontsize=14)
plt.xlabel('요일')
plt.ylabel('평균 매출수량')
plt.tight_layout()
plt.show()

# ------------------------------
# 📈 공휴일 여부 시각화
# ------------------------------
plt.figure(figsize=(6, 4))
sns.barplot(x=holiday_avg.index, y=holiday_avg.values, palette='Set2')
plt.title('공휴일 여부에 따른 평균 매출수량', fontsize=14)
plt.xlabel('공휴일 여부')
plt.ylabel('평균 매출수량')
plt.tight_layout()
plt.show()
```
